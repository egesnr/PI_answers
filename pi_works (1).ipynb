```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/country_vaccination_stats.csv")
```

```python
#Question 4

for country, group in df.groupby('country'):
    df.loc[group.index, 'daily_vaccinations'] = group['daily_vaccinations'].fillna(group['daily_vaccinations'].min())

df['daily_vaccinations'].fillna(0, inplace=True)
```

```python
df.groupby("country", as_index=False).daily_vaccinations.min()
```

```python
#Question 5


top3 = df.groupby("country", as_index=False).daily_vaccinations.mean().sort_values('daily_vaccinations', ascending=False)[0:3]
top3
```

```python
#Questions 6
sum =  df[df['date']=='1/6/2021']['daily_vaccinations'].sum()
sum
```

```python
#Question 7

UPDATE daily_vaccinations_table
SET daily_vaccinations = COALESCE(
    (
        SELECT PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY daily_vaccinations) OVER (PARTITION BY country)
        FROM daily_vaccinations_table sub 
        WHERE sub.country = daily_vaccinations_table.country AND daily_vaccinations IS NOT NULL
    ), 0)
WHERE daily_vaccinations IS NULL;
```
